```python
import spacy

from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Tungsten is a rare metal found naturally on Earth almost exclusively as compounds with other elements. It was identified as a new element in 1781."
doc = nlp(text)
```

```python
displacy.render(doc, style="ent", jupyter=True, options={'distance': 100})
```

```python
sentence_spans = list(doc.sents)
displacy.render(sentence_spans, style="dep", jupyter=True, options={'distance': 80})
```

```python
nlp = spacy.blank("en")
doc = nlp(text)

from spacy.tokens import Span

doc.spans["sc"] = [
    Span(doc, 3, 6, "LOC"),
    Span(doc, 5, 6, "DATE"),
]
displacy.render(doc, style="span", jupyter=True, options={'distance': 100})
```
